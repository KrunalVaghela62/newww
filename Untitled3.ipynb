```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data (V_D and I_D1)
V_D = np.array([0.277, 0.308, 0.315, 0.322, 0.335, 0.347, 0.376, 0.397, 0.45, 0.46,
                0.473, 0.503, 0.506, 0.509, 0.519, 0.523, 0.525, 0.528, 0.53, 0.535,
                0.54, 0.545, 0.548, 0.556, 0.56, 0.564, 0.567, 0.571, 0.575, 0.59,
                0.607, 0.617, 0.65, 0.675, 0.7, 0.725, 0.75, 0.775, 0.8, 0.823])
I_D1 = np.array([1.50E-06, 2.60E-06, 3.00E-06, 3.40E-06, 4.40E-06, 5.50E-06, 1.00E-05, 1.55E-05,
                 4.43E-05, 5.39E-05, 6.97E-05, 1.27E-04, 1.35E-04, 1.45E-04, 1.77E-04, 1.90E-04,
                 1.97E-04, 2.10E-04, 2.22E-04, 2.44E-04, 2.72E-04, 3.01E-04, 3.17E-04, 3.74E-04,
                 4.07E-04, 4.44E-04, 4.73E-04, 5.18E-04, 5.55E-04, 7.62E-04, 1.096E-03, 1.359E-03,
                 2.610E-03, 4.230E-03, 6.870E-03, 1.066E-02, 1.633E-02, 2.430E-02, 3.480E-02,
                 4.980E-02])

# Calculating ln(I_D1)
ln_I_D1 = np.log(I_D1)

# Step 1: Calculate slopes between consecutive points
slopes = np.zeros(len(V_D) - 1)
for i in range(len(V_D) - 1):
    slopes[i] = (ln_I_D1[i+1] - ln_I_D1[i]) / (V_D[i+1] - V_D[i])

# Step 2: Calculate the difference between consecutive slopes
slope_diffs = np.abs(np.diff(slopes))

# Set a threshold for slope difference, based on observation (adjust this threshold if necessary)
threshold = 0.2  # You can adjust this value depending on the data behavior

# Identify the linear region based on slope differences
linear_indices = [0]  # Start with the first point

for i in range(1, len(slope_diffs)):
    if slope_diffs[i-1] < threshold:  # If the change in slope is small, consider this as part of the linear region
        linear_indices.append(i+1)

# Step 3: Use the identified linear indices to fit a line
V_linear = V_D[linear_indices]
ln_I_linear = ln_I_D1[linear_indices]

# Perform linear fit (ln(I_D) = m * V_D + b)
m, b = np.polyfit(V_linear, ln_I_linear, 1)

# Calculate the saturation current (I_S) and ideality factor (n)
n = 1 / (m * 26e-3)  # Assuming V_T = 26mV
I_S = np.exp(b)

# Plot the data and the linear fit
plt.figure(figsize=(8, 6))
plt.plot(V_D, ln_I_D1, 'o', label='Experimental Data', markerfacecolor='blue')  # Open markers (no line)
plt.plot(V_linear, m * V_linear + b, 'r-', label=f'Linear Fit: Slope = {m:.2e}, Intercept = {b:.2e}')
plt.xlabel('V_D (Voltage)')
plt.ylabel('ln(I_D) (Log Current)')
plt.title('Plot of ln(I_D) vs. V_D with Linear Fit')
plt.grid(True)
plt.legend()
plt.show()

# Output the results
print(f"Saturation Current (I_S): {I_S:.2e}")
print(f"Ideality Factor (n): {n:.2f}")
print(np.exp(ln_I_linear))
```
