```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
import cmath

# Step 1: Read the image
image = cv2.imread('ques1.jpg')

# Step 2: Convert the image to grayscale
i_mat= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def dic_con(image):
    image_proj=image.reshape(image.shape[0]*image.shape[1])
    return{"height":image.shape[0],
          "width":image.shape[1],
          "pixels":image_proj}
i=dic_con(i_mat)

def padding(image):
    
    image_0_padded=np.array([0]*(4*i["width"]*i["height"]))
    i_p=image_0_padded.reshape((2*i["height"],2*i["width"]))

    start_row = (i_p.shape[0] - image.shape[0]) // 2
    start_col = (i_p.shape[1] - image.shape[1]) // 2

    end_row = start_row + image.shape[0]
    end_col = start_col + image.shape[1]

    # Place the i_pixels array in the center of image_0_padded
    i_p[start_row:end_row, start_col:end_col] = image
    return i_p
def unpadding(image):
    image_0_padded=np.array([0]*(i["width"]*i["height"]))
    i_p=image_0_padded.reshape((i["height"],i["width"]))

    start_row = (-i_p.shape[0] + image.shape[0]) // 2
    start_col = (-i_p.shape[1] + image.shape[1]) // 2

    end_row = start_row + i_p.shape[0]
    end_col = start_col + i_p.shape[1]

    # Place the i_pixels array in the center of image_0_padded
    i_p=image[start_row:end_row, start_col:end_col]
    return i_p
    
def gaussian_fil(var):
    gaussian_filter = np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            gaussian_filter[k][j] = (1 / (2 * math.pi * (var**2))) * math.exp(-0.5 * (((k-(i["height"]))**2) + ((j-(i["width"]))**2)) / (var**2))
    return gaussian_filter
def lowpass_filter_circular(r):
    lowpass=np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            if (k-i["height"])**2+(j-i["width"])**2<=r**2:
                lowpass[k][j]=1
            else:
                lowpass[k][j]=0
    return lowpass
def lowpass_filter_box(a):
    lowpass=np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            if k in range(i["height"]-a,i["height"]+a+1) and j in range(i["width"]-a,i["width"]+a+1) :
                lowpass[k][j]=1
            else:
                lowpass[k][j]=0
    return lowpass
def butterworth_(D,n):
    butterworth=np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            butterworth[k][j]=1/(1+((((k-i["height"])**2)+((j-i["width"])**2))/(D**2))**n)
    return butterworth
def inv_gaussian_fil(var):
    gaussian_filter = np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            gaussian_filter[k][j] =1-( math.exp(-0.5 * (((k-(i["height"]))**2) + ((j-(i["width"]))**2)) / (var**2)))
    return gaussian_filter
def highpass_filter_circular(r):
    highpass=np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            if (k-i["height"])**2+(j-i["width"])**2<=r**2:
                highpass[k][j]=0
            else:
                highpass[k][j]=1
    return highpass
def highpass_filter_box(a):
    highpass=np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            if k in range(i["height"]-a,i["height"]+a+1) and j in range(i["width"]-a,i["width"]+a+1) :
                highpass[k][j]=0
            else:
                highpass[k][j]=1
    return highpass
def invbutterworth_(D,n):
    butterworth=np.zeros((2*i["height"], 2*i["width"]), dtype=float)
    for k in range(2*i["height"]):
        for j in range(2*i["width"]):
            butterworth[k][j]=1/(1+((D**2)/(((k-i["height"])**2)+((j-i["width"])**2)+1))**n)
    return butterworth
def laplacian_filter1():
    lap=np.zeros((2*i["height"], 2*i["width"]), dtype=complex)
    for k in range(2*i["height"]): 
        for j in range(2*i["width"]):
            lap[k][j] = cmath.exp(1j * 2* 10 * math.pi * j / (2 * i["width"])) - 1  # Use cmath.exp for complex numbers
    return lap
def laplacian_filter2():
    lap=np.zeros((2*i["height"], 2*i["width"]), dtype=complex)
    for k in range(2*i["height"]): 
        for j in range(2*i["width"]):
            lap[k][j] = (2*math.cos( 2* 10 * math.pi * j / (2 * i["width"]))+2*math.cos( 2* 10 * math.pi * k / (2 * i["height"])) - 4 )/4 # Use cmath.exp for complex numbers
    return lap
def power_spectral_density(fourier_image):
    # Assuming z is already defined with shape (6048, 8064)
    # Create x and y as 2D grids matching the shape of z
    z=(abs(np.log1p(fourier_image)))**2
    x = np.arange(z.shape[1])  # Range of columns
    y = np.arange(z.shape[0])  # Range of rows
    x, y = np.meshgrid(x, y)   # Create 2D grids
    # Create the plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the surface
    surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

    # Label axes
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')

    # Show the plot
    plt.show()
    
```

```python
i_pad=padding(i_mat)##Zero pad the image symmetrically to create a new image f’ of size 2H x 2W
##Compute the Fourier transform of f’ and center it using fftshift. Let the Fourier transform be F’(u,v).
transform=np.fft.fft2(i_pad)
transform_shifted=np.fft.fftshift(transform)
fftr=transform_shifted
```

```python
high=gaussian_fil(50)##Design a frequency domain filter H(u,v) of size 2H x 2W with the zero frequency at index (H,W) of the 2D filter array
```

```python
product=fftr*high##Compute the product F’(u,v)H(u,v).
#Compute the inverse Fourier transform of the product after applying ifftshift (to undo the effect of fftshift).
filtered_image = np.fft.ifftshift(product)
filtered_image = np.fft.ifft2(filtered_image)
filtered_image = np.abs(filtered_image)#Consider only the magnitude(we want intensity!!) of the inverse Fourier transform  
#extract the central portion of size H x W.(unpadding)
filtered_image=unpadding(filtered_image)
plt.imshow(filtered_image, cmap='gray')
```

```python
four_mag=(abs(high))**2
con=high.conj()
product_f=(product)*(con/(four_mag+10e-100))#applying weiner filter(choosing approprite noise to signal ratio)
filtered_image_f = np.fft.ifftshift(product_f)
filtered_image_f = np.fft.ifft2(filtered_image_f)
filtered_image_f = np.abs(filtered_image_f)
filtered_image_f=unpadding(filtered_image_f)
plt.imshow(filtered_image_f, cmap='gray')
plt.show()
```
